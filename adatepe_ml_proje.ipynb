```python
!pip install folium
import folium
import requests
import pandas as pd
import time
```

```python
# API Key'iniz (senin verdiğin)
API_KEY = "AIzaSyDR_k6dbVX39MOic3oRgn-d0QhsDAl2ZH8"

# Adatepe Mahallesi sınırları (yaklaşık)
min_lat = min(38.37523196763985, 38.39901486868795, 38.402182362246684, 38.38272270956203)
max_lat = max(38.37523196763985, 38.39901486868795, 38.402182362246684, 38.38272270956203)
min_lon = min(27.192090800626154, 27.2110501568476, 27.18853164418245, 27.173952350726168)
max_lon = max(27.192090800626154, 27.2110501568476, 27.18853164418245, 27.173952350726168)


# Grid hassasiyeti (0.001 derece ≈ 100m, 0.0005 ≈ 50m)
step = 0.0005  # daha sık istersen küçült: 0.00025 → 25m
```

```python
points = []
statuses = []

lat = min_lat
while lat <= max_lat:
    lon = min_lon
    while lon <= max_lon:
        url = f"https://maps.googleapis.com/maps/api/streetview/metadata?location={lat},{lon}&key={API_KEY}"
        response = requests.get(url)
        data = response.json()

        status = data.get("status", "ERROR")
        points.append((lat, lon))
        statuses.append(status)

        print(f"{lat:.6f}, {lon:.6f} → {status}")
        time.sleep(0.1)  # API'yi yormamak için kısa bekleme
        lon += step
    lat += step

# DataFrame oluştur
df = pd.DataFrame(points, columns=["lat", "lon"])
df["status"] = statuses
```

```python
# Harita ortası
map_center = [(min_lat + max_lat) / 2, (min_lon + max_lon) / 2]
m = folium.Map(location=map_center, zoom_start=16)

# Noktaları çiz
for _, row in df.iterrows():
    color = "green" if row["status"] == "OK" else "gray"
    folium.CircleMarker(
        location=(row["lat"], row["lon"]),
        radius=3,
        color=color,
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Haritayı HTML olarak kaydet
m.save("adatepe_streetview_map.html")
print("✅ Harita oluşturuldu: adatepe_streetview_map.html")
```
